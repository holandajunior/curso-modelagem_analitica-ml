```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from IPython.display import Image
```

```python
iris = load_iris()
```

## Usando somente alguns atributos

```python
X = iris.data[:, 2:]
y = iris.target
```

```python
X.shape
```

```python
X[:5,:]
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 5)
tree_clf.fit(X, y)
```

```python
tree_clf.score(X,y)
```

```python
from sklearn.externals.six import StringIO
import pydotplus
from sklearn.tree import export_graphviz
```

```python
dot_data = StringIO()
export_graphviz(tree_clf, out_file=dot_data, feature_names= iris.feature_names[2:], class_names=iris.target_names, rounded = True, filled = True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
X2 = iris.data[:,:]
y2 = iris.target
```

```python
tree_2 = DecisionTreeClassifier()
tree_2.fit(X2, y2)
tree_2.score(X2, y2)
```

```python
dot_data = StringIO()
export_graphviz(tree_2, out_file=dot_data, feature_names= iris.feature_names, class_names=iris.target_names, rounded = True, filled = True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
